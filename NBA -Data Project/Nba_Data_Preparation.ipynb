### Calculating new columns that may help for later analysis

```python
nba['PPG'] = (nba.PTS/nba.GP).round(2)
nba['APG'] = (nba.AST/nba.GP).round(2)
nba['RPG'] = (nba.REB/nba.GP).round(2)
nba['ORPG'] = (nba.ORB/nba.GP).round(2)
nba['DRPG'] = (nba.DRB/nba.GP).round(2)
nba['SPG'] = (nba.STL/nba.GP).round(2)
nba['BPG'] = (nba.BLK/nba.GP).round(2)
nba['FPG'] = (nba.PF/nba.GP).round(2)
nba['MPG'] = (nba.MIN/nba.GP).round(2)
nba['TPG'] = (nba.TOV/nba.GP).round(2)
nba['FT%'] = nba.apply(lambda x: (x['FTM'] / x['FTA']) if x['FTA'] > 15 else np.nan, axis=1)
nba['FG%'] = nba.apply(lambda x: (x['FGM'] / x['FGA']) if x['FGA'] > 15 else np.nan, axis=1)
nba['3P%'] = nba.apply(lambda x: (x['3PM'] / x['3PA']) if x['3PA'] > 15 else np.nan, axis=1)
    
nba.head()
```

```python
print("Top 5 players last 20 years 3P% (atleast 15 shots)")
nba.nlargest(5, '3P%')
```

```python
print("Top 5 players last 20 years FG% (atleast 15 shots)")
nba.nlargest(5, 'FG%')
```

```python
print("Top 5 players last 20 years FT% (atleast 15 shots)")
nba.nlargest(5, 'FT%')
```
